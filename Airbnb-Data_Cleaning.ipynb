```python
columns_to_keep = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
columns_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules','license','id']
```

```python
len(columns_to_keep)
```

```python
len(columns_to_drop)
```

```python
df.drop(columns = columns_to_drop, inplace = True)
```

```python
# Renaming the columns
df.rename(columns={"NAME": "Name"}, inplace = True)
```

```python
# To convert everything in UPPER CASE
new_column_names = []
for i in df.columns:
    new_column_names.append(i.upper())
```

```python
new_column_names
```

```python
new_dataframe_to_modify = df.copy()
```

```python
new_dataframe_to_modify.columns = new_column_names
```

```python
new_dataframe_to_modify.columns
```

```python
new_dataframe_to_modify
```

```python
# Cleaning Duplicates
df.drop_duplicates(inplace= True)
```

```python
df.drop(columns = 'last review', inplace = True)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df['host_identity_verified'] = df['host_identity_verified'].str.upper()
```

```python
df
```

```python
df['instant_bookable'] = df['instant_bookable'].apply(lambda x: 1 if x == True else 0)
```

```python
df
```

```python
df.reset_index(inplace = True)
```

```python
type(df['instant_bookable'][5])
```

```python
df['price'] = df['price'].str.replace("$","")
df['price'] = df['price'].str.replace(",","")
df['price'] = df['price'].str.replace(" ","")
```

```python
type(df['price'][1])
```

```python
df['price'] = df['price'].astype(int)
```

```python
df
```

```python
df['service fee'] = df['service fee'].str.replace("$","")
```

```python
df
```

```python
df.to_csv("Cleaned_Data.csv")
```
